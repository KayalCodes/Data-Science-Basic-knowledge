```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def step_function(x):
    return np.where(x >= 0, 1, 0)

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

def tanh_function(x):
    return np.tanh(x)

def relu_function(x):
    return np.maximum(0, x)

def leaky_relu_function(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def elu_function(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Generate input values
x = np.linspace(-10, 10, 400)

# Compute the function values
step_y = step_function(x)
sigmoid_y = sigmoid_function(x)
tanh_y = tanh_function(x)
relu_y = relu_function(x)
leaky_relu_y = leaky_relu_function(x)
elu_y = elu_function(x)

# Plot the activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, step_y, label='Step Function')
plt.title('Step Function')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, sigmoid_y, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, tanh_y, label='Tanh Function')
plt.title('Tanh Function')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, relu_y, label='ReLU Function')
plt.title('ReLU Function')
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(x, leaky_relu_y, label='Leaky ReLU Function')
plt.title('Leaky ReLU Function')
plt.grid(True)

plt.subplot(2, 3, 6)
plt.plot(x, elu_y, label='ELU Function')
plt.title('ELU Function')
plt.grid(True)

plt.tight_layout()
plt.show()
```
